6. Замість функції з п. 4 створіть генератор, який буде визначати послідовність книжкових сторінок для кожного зошита. Ці рядки можна потім вивести або у циклі, або за допомогою функції next(). Функція-декоратор має працювати з цим генератором так само, як і зі звичайною функцією.

```python
def getvalue():
    while 1:
        x = input("Введіть кількість сторінок:")
        try:
            value = int(x)
            if (value % 16 == 0) and ( 16 <= value <= 1280):
                return value
            else: 
                raise ValueError
        except:
            break
        
def decor(arg):
    def wrap(func):
        def wrapper(num):
            if not arg:
                for i in func(num):
                    print(*i)
            else:
                print(func(num))
        return wrapper
    return wrap

@decor(True)
def func(number_pages):
    def numbs(y):
        lstnumbs = []
        for i in range(1, 16//2+1):
            if i % 2 == 0:
                lstnumbs += [i + y, 16 - i + 1 + y] if i % 2 == 0 else [16 - i + 1 + y, i + y]
        return lstnumbs
    result = []
    for k in range(number_pages // 16):
        result += [numbs(k * 16)]
    return result
    
number_pages = getvalue()
func(number_pages)
```
